```python
# Copyright 2020 aaaaaaaalesha

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CompleteDataset.csv", low_memory=False)

r_df = df.groupby("Nationality").get_group("Russia")
```

```python
r_df # Остались только футболисты из России
```

```python
r_df.groupby("Club").size().plot(kind="barh", figsize=(16,9), grid=True, color="red", edgecolor="black")

plt.title("Принадлежность русских игроков к спортивным клубам", fontdict={'fontsize': 16})
plt.xlabel("Количесво футболистов", fontdict={'fontsize': 14})
plt.ylabel("Клуб", fontdict={'fontsize': 14})
plt.xticks(range(0, 23, 2))

plt.savefig('results_viz3.png')
```

### Текстовое описание визуализации

Данная горизонтальная столбчатая диаграмма показывает распределение игроков из России по футбольным клубов.

Можно увидеть, что:
    
- в командах Зенит, Динамо, Арсенал в полном составе играют русские игроки;
- всего в игре существует 27 команд, в которых присутствует **хотя бы один** русский;
- суммарно в игре 306 русских игроков, из которых 16 играют за Спартак; ¯\_(ツ)_/¯
